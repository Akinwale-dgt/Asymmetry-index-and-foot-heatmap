```python
import pandas as pd
import matplotlib.pyplot as plt

# Function to compute Asymmetry Index for each row and then find the average
def compute_asymmetry_index(file_path, subtask_name):
    # Load the data from the Excel file
    data = pd.read_excel(file_path)

    # Rename the columns for clarity (assuming 1st column is right, 2nd column is left)
    data.columns = ['Right Value', 'Left Value']

    # Calculate the Asymmetry Index for each row
    data['Asymmetry Index'] = abs(data['Right Value'] - data['Left Value']) / ((data['Right Value'] + data['Left Value']) / 2) * 100

    # Calculate the average Asymmetry Index
    avg_asymmetry_index = data['Asymmetry Index'].mean()

    # Return the subtask name and the average asymmetry index
    return subtask_name, avg_asymmetry_index

# Function to process all subtasks
def process_all_subtasks(file_paths):
    subtask_names = ['Measuring', 'Assembly', 'Nailing', 'Lifting', 'Moving', 'Installing']
    results = []

    # Loop through each subtask and file path, calculate the Asymmetry Index
    for i, file_path in enumerate(file_paths):
        subtask_name, ai = compute_asymmetry_index(file_path, subtask_names[i])
        results.append((subtask_name, ai))

    # Return the results as a DataFrame for easier display
    results_df = pd.DataFrame(results, columns=['Subtask', 'Asymmetry Index (%)'])
    return results_df

# Example file paths for each subtask (replace with actual file paths)
file_paths = [
    '/content/drive/MyDrive/Pressure distribution/ASY/MAXPG/MEASURE.xlsx',    # Replace with actual path for Measuring
    '/content/drive/MyDrive/Pressure distribution/ASY/MAXPG/assembly.xlsx',     # Replace with actual path for Assembly
    '/content/drive/MyDrive/Pressure distribution/ASY/MAXPG/nail.xlsx',      # Replace with actual path for Nailing
    '/content/drive/MyDrive/Pressure distribution/ASY/MAXPG/lift.xlsx',      # Replace with actual path for Lifting
    '/content/drive/MyDrive/Pressure distribution/ASY/MAXPG/move.xlsx',       # Replace with actual path for Moving
    '/content/drive/MyDrive/Pressure distribution/ASY/MAXPG/install.xlsx'    # Replace with actual path for Installing
]

# Process all subtasks and get the results
subtask_results = process_all_subtasks(file_paths)

# Plotting the results on a line graph
plt.figure(figsize=(10, 6))
plt.plot(subtask_results['Subtask'], subtask_results['Asymmetry Index (%)'], marker='o', linestyle='-', color='b')

# Set the x and y axis limits to start from 0
plt.ylim(bottom=0)

# Adding title and labels with DejaVu Serif font
plt.xlabel('Subtasks', fontsize=12, fontname='DejaVu Serif')
plt.ylabel('Asymmetry Index (%)', fontsize=12, fontname='DejaVu Serif')

# Update tick labels to DejaVu Serif font
plt.xticks(fontsize=10, fontname='DejaVu Serif')
plt.yticks(fontsize=10, fontname='DejaVu Serif')

# Adding grid
plt.grid(True)

# Display the graph
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import spearmanr

# Load the Excel file
file_path = '/content/drive/MyDrive/Pressure distribution/corr.xlsx'
df = pd.read_excel(file_path)

# Define the columns for perceived rates of discomfort (PRD) and Asymmetric Index (ASI)
prd_columns = df.columns[:9]
asi_columns = df.columns[9:]

# Handle NaN and infinite values by dropping rows with such values
df = df.replace([float('inf'), float('-inf')], float('nan')).dropna()

# Create scatter plots and calculate Spearman correlation coefficients
for prd in prd_columns:
    for asi in asi_columns:
        plt.figure(figsize=(3, 3))  # Adjusted figure size to fit 4 plots side by side on A4 paper
        sns.regplot(x=df[prd], y=df[asi], ci=None, line_kws={"color": "red"}, scatter_kws={"color": "purple"})
        plt.xlabel(prd, fontsize=10)
        plt.ylabel(f'ASI of {asi}', fontsize=10)
        #plt.title(f'Scatter plot between {prd} and ASI of {asi}', fontsize=10)

        # Calculate Spearman correlation coefficient and p-value
        corr_coef, p_value = spearmanr(df[prd], df[asi])
        plt.annotate(f': {corr_coef:.2f}, p: {p_value:.4f}', xy=(0.05, 0.95), xycoords='axes fraction', fontsize=10, color='blue')

        # Adjust layout for better readability
        plt.tight_layout()

        # Show plot
        plt.show()


```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import spearmanr
from matplotlib.backends.backend_pdf import PdfPages
from matplotlib import rcParams

# Set font to DejaVu Serif globally
rcParams['font.family'] = 'DejaVu Serif'

# Load the Excel file
file_path = '/content/drive/MyDrive/Pressure distribution/corr.xlsx'
df = pd.read_excel(file_path)

# Define the columns for perceived rates of discomfort (PRD) and Asymmetric Index (ASI)
prd_columns = df.columns[1:9]  # Skip the first column for overall PRD
asi_columns = df.columns[9:]

# Handle NaN and infinite values by dropping rows with such values
df = df.replace([float('inf'), float('-inf')], float('nan')).dropna()

# Create a PDF to save the plots
output_pdf_path = '/content/drive/MyDrive/Pressure distribution/CORRELATION.xlsx'
with PdfPages(output_pdf_path) as pdf:
    for asi in asi_columns:
        fig, axs = plt.subplots(2, 4, figsize=(12, 6))  # 2 rows, 4 columns for 8 graphs
        axs = axs.flatten()

        for i, prd in enumerate(prd_columns):
            # Plot scatter and regression
            sns.regplot(
                x=df[prd], y=df[asi], ci=None,
                line_kws={"color": "red"}, scatter_kws={"color": "purple"},
                ax=axs[i]
            )
            axs[i].set_xlabel(prd, fontsize=10)
            axs[i].set_ylabel(f'ASI of {asi}', fontsize=10)

            # Calculate Spearman correlation coefficient and p-value
            corr_coef, p_value = spearmanr(df[prd], df[asi])
            axs[i].annotate(
                f': {corr_coef:.2f}, p: {p_value:.4f}',
                xy=(0.05, 0.95), xycoords='axes fraction',
                fontsize=10, color='blue'
            )

        # Adjust layout
        plt.tight_layout()

        # Save page to PDF
        pdf.savefig(fig)
        plt.close(fig)

print(f"Plots have been saved to {output_pdf_path}")
```
